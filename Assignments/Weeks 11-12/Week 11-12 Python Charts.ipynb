```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading data into dataframe.
edu_df = pd.read_csv('ex6-2/education.csv')

# Viewing the dataframe.
edu_df.head()
```

```python
# Plotting histogram.
edu_df['percent_graduates_sat'].hist(grid = False)
plt.title('SAT Graduates percentage of all states')
plt.xlabel('Percentage')
plt.ylabel('No. of states with SAT graduates in percentages')
plt.show()
```

```python
# Plotting boxplot.
plt.boxplot(edu_df.iloc[:,1:4])
plt.title('Reading, Math, Writing of all states.')
plt.xticks([1, 2, 3], edu_df.iloc[:,1:4])
plt.ylabel('Scores')
plt.show()
```

```python
# Plotting bullet chart.
limits = [400, 500, 600]
data_to_plot = edu_df.tail(6)

palette = sns.color_palette("Blues_r", len(limits))

for i in data_to_plot.index:
    fig, ax = plt.subplots()
    ax.set_aspect('equal')
    ax.set_yticks([1])
    ax.set_yticklabels([data_to_plot['state'][i]])

    prev_limit = 0
    for idx, lim in enumerate(limits):
        ax.barh([1], lim-prev_limit, left=prev_limit, height=15, color=palette[idx])
        prev_limit = lim
    
    ax.barh([1], data_to_plot['reading'][i], color='black', height=5) 
      
```
