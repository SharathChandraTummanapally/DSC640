```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import squarify
```

```python
#Loading Dateset
exp_df = pd.read_csv("expenditures.txt",  sep='\t')

exp_df
```

```python
# Plotting treemap
exp_df_2000 = exp_df.loc[exp_df.year ==2000]
squarify.plot(sizes=exp_df_2000['expenditure'], label=exp_df_2000['category'], alpha=.8)
plt.axis('off')
plt.title('Expenditures by Category for the Year 2000')
plt.rcParams["figure.figsize"] = (12,10)
plt.show()
```

```python
#Filtering dataset by Apparel Category
exp_app_df = exp_df.loc[exp_df.category == 'Apparel']
exp_app_df
```

```python
# Plotting Area map
plt.fill_between(exp_app_df['year'], exp_app_df['expenditure'], alpha=0.8)
plt.plot(exp_app_df['year'], exp_app_df['expenditure'], alpha=0.6)
plt.title('Expenditures by Year for Apparel Category', fontsize=10)
plt.xlabel('Year')
plt.ylabel('Expenditure')
plt.show()
```

```python
# Plotting Stacked Area map
exp_df.groupby(['year','category'])['expenditure'].sum().unstack().plot(kind='area', stacked=True)
plt.title('Expenditure by Year for all Categories') 
plt.xlabel('Year')
plt.ylabel('Total Expenditure')
plt.legend(bbox_to_anchor = (1, 1))
```
