```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('hotdog-contest-winners.xlsm')

df.head(15)
```

```python
df.plot(kind='bar' , x='Year', y='Dogseaten',title='Dogs eaten by winner every year');
```

```python
df.groupby(['Country','Winner'])['Dogseaten'].sum().unstack().plot(kind='bar', stacked=True)
plt.title('Total Dogs eaten by Different Winners for each country')       
plt.ylabel('Dogs eaten')
plt.legend(bbox_to_anchor = (1, 0.5))
```

```python
df.groupby('Country')['Winner'].count().plot(kind='pie')
plt.title('Total Winners from each country')
plt.ylabel(' ')
```

```python
df.groupby('Country').apply(lambda x: x[x['Newrecord'] == 1]['Newrecord'].count()).plot(kind='pie')
plt.title('Total New Records from each country')
plt.ylabel(' ')
centre_circle = plt.Circle((0, 0), 0.40, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
```
