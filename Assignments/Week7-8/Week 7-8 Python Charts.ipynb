```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
```

```python
# Loading data to dataframe.
CrimeRatesByState_df = pd.read_csv('ex4-2/crimerates-by-state-2005.csv')

# Viewing the dataframe.
CrimeRatesByState_df.head()
```

```python
# Checking dataframe rows and columns.
CrimeRatesByState_df.shape
```

```python
# Dropping United states and District of columbia from dataframe.
CrimeRatesByState_df.drop(CrimeRatesByState_df[(CrimeRatesByState_df['state'] == 'United States') | (CrimeRatesByState_df['state'] == 'District of Columbia')].index
, inplace = True)
```

```python
# Re-Checking dataframe rows and columns.
CrimeRatesByState_df.shape
```

```python
# Creating a scatterplot.
plt.scatter(CrimeRatesByState_df['burglary'], CrimeRatesByState_df['larceny_theft'])
plt.title('Burglary vs. Larceny theft in the U.S.')
plt.xlabel('Burglary per 100,000 Population')
plt.ylabel('Larceny theft per 100,000 Population')
```

```python
# Using the scatterplot function to build the bubble map
sb.scatterplot(data=CrimeRatesByState_df, x='burglary', y='larceny_theft', size='state', legend=False, sizes=(5, 200))
plt.title('Burglary vs. Larceny theft in the U.S.')
plt.xlabel('Burglary per 100,000 Population')
plt.ylabel('Larceny theft per 100,000 Population')
```

```python
# Creating a densityplot.
CrimeRatesByState_df['burglary'].plot.density(color='green')
plt.title('Burglary rate in the U.S.')
plt.show()
```
